```python
from embedding import HubbardEmbedding
import torch
import einops as ein
import torch.nn.functional as F
```

```python
from occupation_embedding import HubbardEmbedding

token_dims = 32
param_dims = 16
n_params = 4
max_len = 100
test_len = 30
possible_occupations = 5
possible_spin_states = 2
batch_size = 18
```

```python
he = HubbardEmbedding(
    token_dims=token_dims,
    param_dims=param_dims,
    n_params=n_params,
    max_len=max_len,
    possible_occupations=possible_occupations,
    possible_spin_states=possible_spin_states,
)
```

```python
dummy_occupations = torch.randint(
    0, possible_occupations, (test_len, batch_size, possible_spin_states)
)

dummy_occupations = F.one_hot(
    dummy_occupations,
    num_classes=possible_occupations,
)

# The Hubbard embedding takes (seq, batch, occ, spin)
dummy_occupations = ein.rearrange(dummy_occupations, "s b sp o-> s b o sp")

# Only one occupation number is filled in
assert torch.all(dummy_occupations.sum(-2) == 1)
```

```python
dummy_params = torch.randn(n_params, batch_size, dtype=torch.complex64)
```

```python
dummy_occupations.shape  # (s, b, o, sp)
```

```python
dummy_occupations = dummy_occupations.to(torch.complex64)
dummy_occupations
```
